```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating dataset for Linear Regression
np.random.seed(42) # Reproducibility
# Feature
X = np.random.uniform(0, 100, 500)
# Target (Linear relation + noise)
Y = 5 * X + 20 + np.random.normal(0, 25, 500)
```

```python
#Function for estimated Y
def predicted(m, X, c):
    Y_pred = m * X + c
    return Y_pred
```

```python
#Function for calculation loss 
def cost(Y,m,X,c) :
    Y_pred = predicted(m,X,c)
    cost = (np.sum( (Y_pred - Y)**2)) / (2*len(Y))
    return cost
```

```python
#Gradien Function
def gradient (Y,m,X,c):
    Y_pred = predicted(m,X,c)
    dm = np.mean((-Y+ Y_pred)*X)
    dc = np.mean((-Y+Y_pred))
    return dm,dc
```

```python
# This function performs gradient descent
def gradDescent(iterations, learn_rate, Y, m, X, c):
    
    #storing values of d_m, d_c and loss to see the trend later on
    del_m = []
    del_c = []
    cost_func = []

    #iterate multiple times to get hhigher accuracy
    while iterations != 0:
        d_m, d_c = gradient(Y, m, X, c)
        
        m -= learn_rate * d_m
        c -= learn_rate * d_c
        
        del_m.append(m)
        del_c.append(c)
        cost_func.append(cost(Y, m, X, c))
        
        iterations -= 1
        
    return m, c, del_m, del_c, cost_func
```

```python
# Train model defining learning rate and no. of iterations
iterations = 1000
learn_rate = 0.00001

m = 0
c = 0

m, c, del_m, del_c, loss_func = gradDescent(iterations, learn_rate, Y, m, X, c)
```

```python
print(m,c)
```

```python
import matplotlib.pyplot as plt
import numpy as np


Y_pred = predicted(m, X, c)

idx = np.argsort(X)

plt.figure(figsize=(8,5))


plt.scatter(X, Y, label="Actual Y")

plt.plot(X[idx], Y_pred[idx], label="Predicted Y",color="red")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression From Scratch")
plt.legend()

plt.show()
plt.savefig("graph.png")
```

```python
from sklearn.metrics import r2_score

Y_pred = predicted(m, X, c)

print("R² Score:", r2_score(Y, Y_pred))
```

```python
plt.plot(loss_func)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function vs Iterations")
plt.show()
```

```python
plt.plot(del_c)
plt.xlabel('No. of iterations')
plt.ylabel('Value of c')
plt.title('Variation in value of Intercept (c)')
plt.show
```

```python

np.random.seed(42)

K = np.random.randint(0, 100, 700)

# True equation: y = 16x + 50 + moderate noise
L = 16 * K + 50 + np.random.normal(0, 200, 700)
```

```python
iterations = 1000
learn_rate = 0.0001

m = 0
c = 0

m, c, del_m, del_c, loss_func = gradDescent(iterations, learn_rate, L, m, K, c)
```

```python
print(m,c)
```

```python
import matplotlib.pyplot as plt
import numpy as np


L_pred = predicted(m, K, c)

idx = np.argsort(X)

plt.figure(figsize=(8,5))


plt.scatter(K, L, label="Actual L")

plt.plot(K[idx], L_pred[idx], label="Predicted L",color="red")

plt.xlabel("K")
plt.ylabel("L")
plt.title("Linear Regression From Scratch")
plt.legend()

plt.show()
```

```python
from sklearn.metrics import r2_score

Y_pred = predicted(m, X, c)

print("R² Score:", r2_score(L, L_pred))
```
